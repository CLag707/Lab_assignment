# Calculate

* $\frac{\partial J}{\partial  \mathbf{W_1}} = \left( \mathbf{X}^T \cdot \sigma'(Z_1) \cdot \left( \frac{1}{m} (\hat{y} - y) \cdot \sigma'(Z_2) \right) \cdot \mathbf{W_2}^T \right) + \lambda \mathbf{W_1}$ 

    *  $\frac{\partial J}{\partial  \mathbf{W_1}}$ = $\frac{\partial L}{\partial  \mathbf{W_1}}$ + $\frac{\partial R}{\partial  \mathbf{W_1}}$

        * $\frac{\partial R}{\partial \mathbf{W_1}}$ = $\lambda\mathbf{W_1}$

        * $\frac{\partial L}{\partial \mathbf{W_1}} = \frac{\partial\mathbf{Z_1}}{\partial \mathbf{W_1}} \cdot \frac{\partial\mathbf{A_1}}{\partial \mathbf{Z_1}} \cdot \frac{\partial\mathbf{Z_2}}{\partial \mathbf{A_1}} \cdot \frac{\partial\mathbf{A_2}}{\partial \mathbf{Z_2}} \cdot \frac{\partial\mathbf{\hat{y}}}{\partial \mathbf{A_2}} \cdot \frac{\partial\ L}{\partial \mathbf{\hat{y}}}$

        * $\frac{\partial L}{\partial \mathbf{W_1}}= \mathbf{X}^T \cdot \sigma^\prime(Z_1) \cdot \mathbf{W_2}^T \cdot \sigma^\prime(Z_2) \cdot-\frac{1}{m}(y-\hat{y})$

<br  />

* $\frac{\partial J}{\partial  \mathbf{W_2}} = \mathbf{A_1}^T \cdot \left( \sigma'(Z_2) \cdot \frac{1}{m} (\hat{y} - y) \right)$

    * $\frac{\partial J}{\partial  \mathbf{W_2}} =\frac{\partial L}{\partial  \mathbf{W_2}} + \frac{\partial R}{\partial  \mathbf{W_2}}$

        * $\frac{\partial R}{\partial \mathbf{W_2}}$ = $\lambda\mathbf{W_2}$

        * $\frac{\partial L}{\partial \mathbf{W_2}} = \frac{\partial\mathbf{Z_2}}{\partial \mathbf{W_2}} \cdot \frac{\partial\mathbf{A_2}}{\partial \mathbf{Z_2}} \cdot \frac{\partial\mathbf{\hat{y}}}{\partial \mathbf{A_2}} \cdot \frac{\partial\ L}{\partial \mathbf{\hat{y}}}$

        * $\frac{\partial L}{\partial \mathbf{W_2}} = \mathbf{A_1}^T \cdot \sigma^\prime(Z_2) \cdot-\frac{1}{m}(y-\hat{y})$

<br  />

* $\frac{\partial J}{\partial \mathbf{b_1}} = \sigma'(Z_1) \cdot \left( \sigma'(Z_2) \cdot \frac{1}{m} (\hat{y} - y) \right) \cdot \mathbf{W_2}^T$

    * $\frac{\partial J}{\partial  \mathbf{b_1}} =\frac{\partial L}{\partial  \mathbf{b_1}} + \frac{\partial R}{\partial  \mathbf{b_1}}$

        * $\frac{\partial R}{\partial \mathbf{b_1}} = 0$

        * $\frac{\partial L}{\partial \mathbf{b_1}} = \frac{\partial\mathbf{Z_1}}{\partial \mathbf{b_1}} \cdot \frac{\partial\mathbf{A_1}}{\partial \mathbf{Z_1}} \cdot \frac{\partial\mathbf{Z_2}}{\partial \mathbf{A_1}} \cdot \frac{\partial\mathbf{A_2}}{\partial \mathbf{Z_2}} \cdot \frac{\partial\mathbf{\hat{y}}}{\partial \mathbf{A_2}} \cdot \frac{\partial\ L}{\partial \mathbf{\hat{y}}}$

        * $\frac{\partial L}{\partial \mathbf{b_1}} =\biggl(\sigma^\prime(Z_1) \cdot \mathbf{W_2}^T \cdot \sigma^\prime(Z_2) \cdot-\frac{1}{m}(y-\hat{y})\biggl)$ 

<br  />

* $\frac{\partial J}{\partial  \mathbf{b_2}} = \sigma^\prime(Z_2) \cdot\frac{1}{m}(\hat{y}-y)$

    * $\frac{\partial J}{\partial  \mathbf{b_2}} =\frac{\partial L}{\partial  \mathbf{b_2}} + \frac{\partial R}{\partial  \mathbf{b_2}}$

        * $\frac{\partial R}{\partial \mathbf{b_2}} = 0$

        * $\frac{\partial L}{\partial \mathbf{b_2}} = \sigma^\prime(Z_2) \cdot-\frac{1}{m}(y-\hat{y})$

# 3: Training ANN

```python
def sigmoid(s):
    return 1/(1+np.exp(-s))

def sigmoid_derivative(s):
    return sigmoid(s) * (1 - sigmoid(s))

def forward(X, W1, b1, W2, b2):
    Z1 = X @ W1 + b1  # (m, 2) dot (2, 2) + (1, 2) = (m, 2)
    A1 = sigmoid(Z1)  # (m, 2)
    Z2 = A1 @ W2 + b2  # (m, 2) dot (2, 1) + (1, 1) = (m, 1)
    A2 = sigmoid(Z2)  # (m, 1)
    y_hat = A2
    return A1, y_hat

def get_cost(y, y_hat, W1, W2, LAMBDA):
    m = len(y)
    error = y - y_hat
    L = np.mean((y-y_hat) ** 2) * .5

    f_W1 = np.linalg.norm(W1, 'fro') # frobenius norm of W1 and W2
    f_W2 = np.linalg.norm(W2, 'fro')
    R = (LAMBDA/2) * ((f_W1**2 + f_W2**2))
    J = L+R
    return J

def get_classification_error_rate(y, y_hat):
    y_binary = (y_hat > 0.5).astype(int)
    error_rate = np.mean(y_binary != y) * 100.0
    return error_rate

def iterate(X, y, W1, b1, W2, b2, ALPHA, LAMBDA):
    m = len(X)
    
    Z1 = X @ W1 + b1  # (m, 2) dot (2, 2) + (1, 2) = (m, 2)
    A1 = sigmoid(Z1)  # (m, 2)
    Z2 = A1 @ W2 + b2  # (m, 2) dot (2, 1) + (1, 1) = (m, 1)
    A2 = sigmoid(Z2)  # (m, 1)
    y_hat = A2
    
    cost = get_cost(y, y_hat, W1, W2, LAMBDA)
    error = get_classification_error_rate(y, y_hat)
    
    gradient_Z2 = (1/m) * (y_hat - y) * sigmoid_derivative(Z2)  # (m, 1)
    gradient_W2 = (A1.T @ gradient_Z2) + LAMBDA * W2  # (2, 1)



    gradient_b2 = np.sum(gradient_Z2, axis=0, keepdims=True)  # (1, 1)
    gradient_A1 = gradient_Z2 @ W2.T  # (m, 1) dot (1, 2) = (m, 2)
    gradient_Z1 = gradient_A1 * sigmoid_derivative(Z1)  # (m, 2)
    gradient_W1 = (X.T @ gradient_Z1) + LAMBDA * W1  # (2, m) dot (m, 2) = (2, 2)
    gradient_b1 = np.sum(gradient_Z1, axis=0, keepdims=True)  # (1, 2)
    
    
    W1 -= ALPHA * gradient_W1
    W2 -= ALPHA * gradient_W2
    b1 -= ALPHA * gradient_b1
    b2 -= ALPHA * gradient_b2
    
    return W1, b1, W2, b2, cost, error

def fit_3_layer_MLP(X, y, MLP_hidden_size=2, initial_lr= 100, lr_decay_rate=0.1, max_epochs= 10000,
                    stopping_threshold=1e-10, l2_lambda=0.0001, print_interval=500, verbosity=1):

    # Dimensions
    inputD = 2  
    hiddenD = 2  
    outputD = 1

    # Weights and biases
    W1 = np.random.randn(inputD, hiddenD) * 0.1  # (2, 2)
    W2 = np.random.randn(hiddenD, outputD) * 0.1  # (2, 1)
    b1 = np.zeros((1, hiddenD))  # (1, 2)
    b2 = np.zeros((1, outputD))  # (1, 1)
    
    costs = []
    errors = []
    
    for epoch in range(max_epochs):
        lr = initial_lr / (1 + lr_decay_rate * (epoch+1))  # Learning decay rate
        
        W1, b1, W2, b2, cost, error_rate = iterate(X, y, W1, b1, W2, b2, lr, l2_lambda)
        
        if epoch % print_interval == 0:
            print(f"Epoch {epoch}/{max_epochs}, Cost: {cost:.9f}, Error: {error_rate:.2f}%")
        
        costs.append(cost)
        errors.append(error_rate)
        
        if cost < stopping_threshold:  # Termination requirement
            print(f"Stopped at epoch {epoch}")
            print(f"Cost: {cost:.4f}")
            break

    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 2, 1)
    plt.plot(costs)
    plt.xlabel('Epoch')
    plt.ylabel('Cost')
    plt.title('Cost over time')

    plt.subplot(1, 2, 2)
    plt.plot(errors)
    plt.xlabel('Epoch')
    plt.ylabel('Error rate (%)')
    plt.title('Error rate over time')

    plt.tight_layout()
    plt.show()
    
    return W1, b1, W2, b2, costs, errors

W1_test, b1_test, W2_test, b2_test, costs, errors = fit_3_layer_MLP(train1, y_train1)
```

# 5 & 6: Visualizing decision boundaries for hidden nodes

```python
def visualize_decision_boundary(neuron_layer, neuron_number, title,
                                X_data, y_data, W1, b1, W2, b2,
                                plot_datapoints=False, show_rounded=False, granularity=500):
    
    # create the meshgrid
    col_maxes = X_data.max(axis=0)
    col_mins = X_data.min(axis=0)
    x0_space = np.linspace(start=col_mins[0], stop=col_maxes[0], num=granularity)
    x1_space = np.linspace(start=col_mins[1], stop=col_maxes[1], num=granularity)
    x0, x1 = np.meshgrid(x0_space, x1_space)
    x0 = x0.flatten()
    x1 = x1.flatten()
    X_grid = np.column_stack((x0,x1))

    # get activations for appropriate neuron and organize into grid
    A = forward(X_grid, W1, b1, W2, b2)

    # get appropriate neuron and round if specified
    activations = A[neuron_layer][:,neuron_number]
    if show_rounded: activations = np.round(activations)
    activations_grid = np.flip(activations.reshape(x1_space.shape[0], x0_space.shape[0]), axis=0)

    # plot
    fig, ax = plt.subplots()
    ax.imshow(activations_grid, vmin=0.5, vmax=1, extent=[col_mins[0]-0.1, col_maxes[0]+0.1, col_mins[1]-0.1, col_maxes[1]+0.1])
    if plot_datapoints:
        positives_mask = np.squeeze(y_data) == 1
        negatives_mask = np.squeeze(y_data) == 0
        ax.scatter(X_data[positives_mask,0], X_data[positives_mask,1], c="orange", label="y = 1")
        ax.scatter(X_data[negatives_mask,0], X_data[negatives_mask,1], c="blue", label="y = 0")
        ax.set_xlabel("x0")
        ax.set_ylabel("x1")
        ax.legend(loc="upper right", bbox_to_anchor=(1.27,1))
    ax.set_title(title)
    plt.show()

visualize_decision_boundary(0, 0, "Decision boundary for the first neuron in the hidden layer", train1, y_train1, W1_test, b1_test, W2_test, b2_test)
visualize_decision_boundary(0, 1, "Decision boundary for the second neuron in the hidden layer", train1, y_train1, W1_test, b1_test, W2_test, b2_test)
```
